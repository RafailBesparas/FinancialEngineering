```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
#plt.style.use("seaborn")
```

Import the stocks

We use header 0,1 in order to show both the Open Price for all stocks and the Stock names then we transform the dataframe


```python
stocks = pd.read_csv("us_stocks.csv", header = [0,1], index_col= [0], parse_dates= [0])
```

```python
stocks.head()
```

Select the ADJ close for all stocks from 31-12-2015 till 31-12-2018? How many Timestamps/days are in the dataframe now?

```python
stocks = stocks.loc["31-12-2015":"31-12-2018", "Adj Close"]
```

```python
stocks.head()
```

```python
stocks.info()
```

## Stock Visualization

```python
stocks.plot(figsize = (12,8))
plt.show()
```

#### Same Base Value 100 Visualization
Bring the stocks to have the same base value of 100

```python
stocks.div(stocks.iloc[0]).mul(100).plot(figsize = (12,8))
plt.show()
```

Comment: Still it is very hard to check the performance of stocks

Resample stocks to monthly frequency
(last traiding price for each month and the index should show the last calendar day)

```python
# Use the resample method to get monthly data
stocks_m = stocks.resample("ME").last()
```

```python
stocks_m.head()
```

Calculate monthly returns and drop nan values

```python
ret = stocks_m.pct_change().dropna()
```

```python
ret
```

Calculate the mean and standard deviation of monthly returns

```python
summary = ret.agg(["mean", "std"]).T
```

```python
summary
```

Make the calculation annually for the mean and standrd deviation for returns

```python
summary["mean"] = summary["mean"]*12
summary["std"] = summary["std"]*np.sqrt(12)
```

```python
summary
```

Plot & Compare the performance of the six stocks in a (total) risk / return frameworks

Identify the best performing stocks and the worst performing stock (in terms of risk/return)

```python
summary.plot(kind = "scatter", x = "std", y = "mean", figsize = (12, 8), s = 50, fontsize = 15, xlim = (0.1, 0.3), ylim = (0, 0.25))
for i in summary.index:
    plt.annotate(i, xy=(summary.loc[i, "std"]+0.002, summary.loc[i, "mean"]+0.002), size = 15)
plt.xlabel("ann. Risk(std)", fontsize = 15)
plt.ylabel("ann. Return", fontsize = 15)
plt.title("Risk/Return", fontsize = 20)
plt.show()
```

With the risk / return framework I can identify the McDonalds has the best performaance while Facebook has the worst
